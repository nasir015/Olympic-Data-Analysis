## Exploratory Data Analysis of olympic data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### import data

```python
df1 = pd.read_csv('athlete_events.csv')
df2 = pd.read_csv('noc_regions.csv')
```

```python
df = pd.merge(df1, df2, on='NOC', how='left')
df.head()
```

#### Rename columns names

```python
df.rename(columns={'region':'Region','notes':'Notes'}, inplace=True)
df.head()
```

#### How many Athlete come from india?

```python
df[df['Region']== 'India'].count()
```

```python
df[df['Region']== 'Japan'].count()
```

```python
# top 10 countries participating 
a =df.Team.value_counts().sort_values(ascending=False).head(10)
A = a.index
B = a.values
```

```python
plt.figure(figsize= (12,6))
plt.title('Top 10 participating countries')
plt.xlabel('Country')
plt.ylabel('Number of Participants')
sns.barplot(x = A, y = B, palette= 'Set2');
```

```python
# Age distribution of athlete
age = df.Age
age
```

```python
age.index
```

```python
plt.figure(figsize= (6,3))
plt.title('Top 5 participating Age groups')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(age,bins= np.arange(10,80,2),color='orange',edgecolor = 'white')
```

```python
#Total  medal in olympic
df.Medal.value_counts()
```

```python
# Heights gold medal takers countries
gold = df[df.Medal== 'Gold'].Region.value_counts().head(10)
```

```python
plt.figure(figsize= (8,5))
plt.title ('Heights gold medal takers countries')
plt.xlabel('countries')
plt.ylabel ('Number of gold medal')
sns.barplot(x =gold.index, y =gold.values, palette= 'rocket');
```

```python
# Total female athlete in each olympic
female_athletes = df[(df.Sex == 'F')&(df.Season == 'Summer') ].groupby(df.Year).size()
female_athletes
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(x = female_athletes.index, y = female_athletes)
```

```python
# Total male athlete in each olympic
male_athletes = df[(df.Sex == 'M')&(df.Season == 'Summer') ].groupby(df.Year).size()
male_athletes
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(x = male_athletes.index, y = male_athletes)
```
